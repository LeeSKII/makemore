```python
import torch
import numpy as np
from matplotlib import pyplot as plt
import torch.nn as nn
import torch.optim as optim
```

```python
# 准备训练数据和测试数据
x_data = np.random.randn(100, 1)
noisy = (np.arange(100)*0.01).reshape(100, 1)
y_data= x_data + noisy
```

```python
# 查看训练数据的形状
plt.scatter(x_data,y_data)
```

```python
# 定义线性回归模型
class LinearRegression(nn.Module):
    def __init__(self):
        super(LinearRegression, self).__init__()
        self.fc = nn.Linear(1,1)
    
    def forward(self, x):
        y_pred = self.fc(x)
        return y_pred
```

```python
# 将训练数据转换成torch的张量
x = torch.tensor(x_data,dtype=torch.float32)
y = torch.tensor(y_data,dtype=torch.float32)
```

```python
# 实例化模型
model = LinearRegression()
```

```python
# 查看模型结构
for name,param in model.named_parameters():
    print(name,param.shape)
```

```python
# 定义模型的优化器和损失函数
optimizer = optim.SGD(model.parameters(),lr = 0.01)
criterion = nn.MSELoss()
```

```python
# 梯度下降迭代训练模型
for i in range(100):
    y_pred = model(x)
    loss = criterion(y_pred, y)
    loss.backward()
    optimizer.step()
    optimizer.zero_grad()
    print(loss.item())
```

```python
# 得到训练完成的模型的预测结果
y_pred = model(x)
```

```python
# 将数据转换成numpy的array
y_pred_arr = y_pred.detach().numpy()
```

```python
# 对比显示线性回归的预测值和真实值
plt.scatter(x_data,y_data)
plt.plot(x_data,y_pred_arr,color='red')
plt.show()
```
